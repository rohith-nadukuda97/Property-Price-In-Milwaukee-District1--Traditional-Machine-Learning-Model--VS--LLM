```python
%pip install requests
%pip install pandas
import requests
import pandas as pd
from io import StringIO

url = "https://data.milwaukee.gov/dataset/5a537f5c-10d7-40a2-9b93-3527a4c89fbd/resource/395db729-a30a-4e53-ab66-faeb5e1899c8/download/wibrarchive.csv"

response = requests.get(url)
if response.status_code == 200:
    
    csv_data = StringIO(response.text)
    df = pd.read_csv(csv_data)

   
    print(df.head())
else:
    print(f"Failed to download the file. Status code: {response.status_code}")

```

```python
data = df[(df['ReportedYear'].isin([2016,2017,2018, 2019, 2020, 2021, 2022])) & (df['POLICE'].isin([1, 5]))]
```

```python
n=pd.DataFrame()
n['crime_count']=data['Arson']+data['AssaultOffense']+data['Burglary']+data['CriminalDamage']+data['Homicide']+data['LockedVehicle']+data['Robbery']+data["SexOffense"]+data['Theft']+data['VehicleTheft']
```

```python
data['crime_count']=n['crime_count']
```

```python
data=data.dropna(subset=['Location'],axis=0)
```

```python

grouped_data = data.groupby(['Location', 'ReportedYear'])['crime_count'].sum().reset_index()




```

```python
grouped_data
```

```python
unique_locations = data['Location'].unique()


print(unique_locations)

# Optionally, save to a new CSV for further processing
unique_location=pd.DataFrame(unique_locations, columns=['Unique Locations'])
```

```python
unique_location
```

```python
from geopy.geocoders import Nominatim
import time

lat = []
long = []
geolocator = Nominatim(user_agent='app',timeout=4)

```

```python
for name in unique_location['Unique Locations'][11800:12700]:
    location = geolocator.geocode(name)
    if location:  # Check if the location was found
        lat.append(location.latitude)  # Correct attribute name
        long.append(location.longitude)  # Correct attribute name
    else:
        lat.append(None)  # Or handle the case as needed
        long.append(None)  # Or handle the case as needed
time.sleep(1) 
```

```python
len(lat)
```

```python
unique_locations_10900=unique_location[0:10900]
```

```python
unique_locations_10900['longitude']=long
unique_locations_10900['latitude']=lat
```

```python
unique_locations_10900
```

```python
unique_locations_10900 = unique_locations_10900.rename(columns={'Unique Locations': 'Location'})
```

```python
merged_df = pd.merge(grouped_data, unique_locations_10900, on='Location', how='left')

print(merged_df)
```

```python
merged_df.isna().sum()
```

```python
merged_df=merged_df.dropna(subset=['longitude'],axis=0)
```

```python
merged_df = merged_df.reset_index(drop=True)
```
