```python

import glob
folder_path = 'Data/*.csv' 
csv_files = glob.glob(folder_path)

dataframes = [pd.read_csv(file) for file in csv_files]
combined_df = pd.concat(dataframes, ignore_index=True)
print("Combined DataFrame:")
print(combined_df.head())
```

changing the Date time Formatinto appropriate format

```python
def convert_date(date):
    try:
        
        pd.to_datetime(date, format='%m/%d/%Y')
        return date  
    except ValueError:
        return pd.to_datetime(date, format='%Y-%m-%d').strftime('%m/%d/%Y')


combined_df['Sale_date'] = combined_df['Sale_date'].apply(convert_date)
```

```python
combined_df.head(5)
```

Cleaning the sale_price column

```python

def clean_price(price):
    try:
        
        if isinstance(price, str):
            return float(price.replace('$', '').replace(',', ''))
        
        return float(price)
    except ValueError:
       
        return price

combined_df['Sale_price'] = combined_df['Sale_price'].apply(clean_price)
combined_df['Lotsize'] = combined_df['Lotsize'].apply(clean_price)


```
